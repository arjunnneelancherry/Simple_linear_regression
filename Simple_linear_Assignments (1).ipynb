```python
time.rename(columns={'Delivery Time':'DeliveryTime','Sorting Time':'SortingTime'},inplace=True)
```

```python
import seaborn as sns
sns.distplot(time['DeliveryTime'])
```

```python
import seaborn as sns
sns.distplot(time['SortingTime'])
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols('SortingTime ~ DeliveryTime',data = time).fit()
```

```python
sns.regplot(x='DeliveryTime',y='SortingTime',data=time)
```

```python
model.params
```

```python
model.fittedvalues
```

```python
time['Predicted']=model.fittedvalues
```

```python
time['Error']=model.resid
```

```python
time['Error_sq']=(time['Error'])**2
time
```

```python
#Intercept      -0.756673
#DeliveryTime    0.413744
time['Y']=-0.756673+(0.413744*time['DeliveryTime'])+time['Error']
time
```

```python
model.rsquared
```

```python
model.pvalues
```

```python
time['log_delivery']=np.log(time['DeliveryTime'])
time
```

```python
model=smf.ols('SortingTime~log_delivery',data=time).fit()
```

```python
model.params
```

```python
time['log_predicted']=model.fittedvalues
time['log_errors']=model.resid
time['Y']=-12.499233+(6.735548*time['log_delivery']+time['log_errors'])
```
